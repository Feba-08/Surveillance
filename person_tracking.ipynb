```python
df=pd.read_csv("C:\DATASCIENCE\Project\person_tracking.csv")
```

```python
df['In_date']=df['In_time'].str.split(expand=True)[0]
```

```python
df['In_time']=df['In_time'].str.split(expand=True)[1]
```

```python
df = df.reindex(['Counter','Elapsed_time','In_time','In_date'], axis=1)
```

```python
df
```

```python
filtered_df = df.loc[(df['In_time'] >= '00:09:00')& (df['In_time'] < '00:12:00')]
```

```python
filtered_df
```

```python
filtered_df = df.loc[(df['Elapsed_time'] >= 50)]
```

```python
filtered_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=filtered_df, x='In_time', y='Elapsed_time',color = 'blue', palette = 'hls')
plt.xticks(rotation=45)
plt.show()
```
